```python
import matplotlib.pyplot as plt

def plot_loss_optimization(loss_list1, loss_list2, label1="Loss 1", label2="Loss 2"):
    """
    Plots the loss optimization process for two loss lists.
    
    Parameters:
    - loss_list1: List of losses for the first process
    - loss_list2: List of losses for the second process
    - label1: Label for the first loss list (default: "Loss 1")
    - label2: Label for the second loss list (default: "Loss 2")
    """
    # Check if the lengths of the lists match
    if len(loss_list1) != len(loss_list2):
        raise ValueError("The two loss lists must have the same length.")
    
    # Create x-axis values (assumed as epochs)
    epochs = range(1, len(loss_list1) + 1)
    
    # Plot the losses
    plt.figure(figsize=(15, 10))
    plt.plot(epochs, loss_list1, label=label1, marker='o', linestyle='-', color='b')
    plt.plot(epochs, loss_list2, label=label2, marker='s', linestyle='--', color='r')
    # plt.ylim(0, 0.05)
    
    # Add title and labels
    plt.title("Loss Optimization Process", fontsize=20)
    plt.xlabel("Epochs", fontsize=15)
    plt.ylabel("Loss", fontsize=15)
    
    # Add legend
    plt.legend(fontsize=15)
    
    # Add grid for better readability
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage.transform import resize

def combine_images_horizontally(image_path1, image_path2, output_path=None):
    """
    Combines two images horizontally after resizing them to the same dimensions.
    Optionally saves the result.

    Parameters:
    - image_path1: Path to the first image
    - image_path2: Path to the second image
    - output_path: Path to save the combined image (optional)
    """
    # Load the images
    img1 = mpimg.imread(image_path1)
    img2 = mpimg.imread(image_path2)

    # Determine the target dimensions (height and width of the smaller image)
    target_height = min(img1.shape[0], img2.shape[0])
    target_width = min(img1.shape[1], img2.shape[1])

    # Resize both images to the target dimensions
    img1_resized = resize(img1, (target_height, target_width), anti_aliasing=True)
    img2_resized = resize(img2, (target_height, target_width), anti_aliasing=True)

    # Combine the images horizontally
    combined_image = np.hstack((img1_resized, img2_resized))

    # Display the combined image
    plt.figure(figsize=(10, 5))
    plt.axis('off')
    plt.imshow(combined_image)
    # plt.title("Combined Image")
    plt.show()

    # Save the combined image if output_path is provided
    if output_path:
        plt.imsave(output_path, combined_image)

# Example usage
image_path1 = "./Q_former.png"
image_path2 = "./joint_loss.png"
output_path = "combined_image.jpg"

combine_images_horizontally(image_path1, image_path2, output_path)
```
